```python
#dealing with multiple inputs
import pandas
from sklearn import linear_model


df = pandas.read_csv("cars.csv")
df.head()

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict([[3300, 1300]])

print(predictedCO2) 
print(regr.coef_) #The coefficient is a factor that describes the relationship with an unknown variable.
```

```python
#train and test (Train/Test is a method to measure the accuracy of your model.)
import numpy
import matplotlib.pyplot as plt

numpy.random.seed(2)
x=numpy.random.normal(3,1,100)
y=numpy.random.normal(150,40,100)/x

train_x=x[:80]#takes the first 80 values from x.Used to train the model (learn patterns).
train_y=y[:80]#Takes the first 80 values from y (corresponding targets).Used as the true answers during training.

test_x=x[80:]
test_y=y[80:]

plt.xlabel('x_normal')
plt.ylabel('y_normal')
plt.scatter(train_x,train_y)
plt.show()
```

```python
#displaying the test data
plt.scatter(test_x,test_y)
plt.show()
```

```python
#draw the line of polynomial regression to fit the data set
mymodel=numpy.poly1d(numpy.polyfit(train_x, train_y, 4))
myline=numpy.linspace(0,6,100)
plt.scatter(train_x, train_y)

plt.xlabel('minutes')
plt.ylabel('money')
plt.plot(myline, mymodel(myline))
plt.show()

```

```python
#if unsure about whether the line fits or not, use r2_score
from sklearn.metrics import r2_score
r2=r2_score(train_y, mymodel(train_x))
print(r2)
print(mymodel(7))

```

```python
#drawing the decision tree

import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("comedy.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()# machine learning algorithm that makes predictions based on a tree of decisions
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features) 
```
