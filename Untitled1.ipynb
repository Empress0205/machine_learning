```python
#confusion matrix It is a table that is used in classification problems to assess where errors in the model were made.The rows represent the actual classes the outcomes should have been. While the columns represent the predictions we have made. Using this table it is easy to see which predictions are wrong.
import numpy
import matplotlib.pyplot as plt
from sklearn import metrics

actual=numpy.random.binomial(1, .9, size=1000)#.binomial(1, .9) means a 90% chance of getting 1 (positive), and 10% for 0 (negative).
predicted=numpy.random.binomial(1, .9, size=1000)

confusion_matrix=metrics.confusion_matrix(actual, predicted)
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=[0,1])
cm_display.plot()
plt.show()

accuracy=metrics.accuracy_score(actual, predicted)
print(accuracy)#How many total predictions were correct?

precision=metrics.precision_score(actual, predicted)
print(precision)#How many predicted positives were actually correct?

Sensitivity_recall = metrics.recall_score(actual, predicted)
print(Sensitivity_recall)#How many actual positives were predicted correctly?

Specificity = metrics.recall_score(actual, predicted, pos_label=0)
print(Specificity)#How many actual negatives (class 0) were predicted correctly?

F1_score = metrics.f1_score(actual, predicted)
print(F1_score)#t balances both in one score  good when classes are imbalanced.
```
