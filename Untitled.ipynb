```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df=pd.read_excel('height_weight_data.xlsx')
df.head()




```

```python
plt.scatter(df['Weight_kg'], df['Height_cm'])
plt.xlabel('Weight_kg')
plt.ylabel('Height_cm')
```

```python
df.corr()
```

```python
##searborn for visualization
import seaborn as sns
sns.pairplot(df)

```

```python
#dependent and independent features
df.head()
x= df[['Weight_kg']] #all independent features should be in form of 2d
y=df['Height_cm']  #dependent features can be in 1d



```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.25, random_state=42)#fot training and testing models
#the 0.25 means 25% of data goes to test while 75% goes to training
```

```python
#standardization(the aim is to standardaze data before training model,ensuring each feature has mean of 0 and standard deviation of 1)
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit_transform(x_train) #fit(x_train calculates mean and standard deviation of the trained data, and transform(x_train) applies transformation
```

```python
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
#apply linear regression (used to predict continuous value basing on one or more input features)
#n_jobs=1 Use all available CPU cores for computation (faster training, especially with large data).
#regression.fit(x_train, y_train)

  #  This line trains the model on your training data:

       # x_train: the input features (independent variables)

       # y_train: the target/label (what you're trying to predict)

#So here, the model learns the relationship between x and y by finding the best-fit line.
from sklearn.linear_model import LinearRegression

regression=LinearRegression(n_jobs=-1)
regression.fit(x_train, y_train)

```

```python
#a coefficient represents the effect of a feature (input variable) on the target (output variable).
#It tells us how much the target value changes when the input feature increases by 1 unit  assuming all other features stay the same.
print("COEFFICIENT OR SLOPE", regression.coef_)
print("intercept is:", regression.intercept_)
```

```python
#plot training data plot best fit line
plt.scatter(x_train, y_train)
plt.plot(x_train, regression.predict(x_train))
```

```python
#prediction for test data
regression.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mse=mean_squared_error(y_test, regression.predict(x_test))
mae=mean_absolute_error(y_test, regression.predict(x_test))
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)


```

```python
from sklearn.metrics import r2_score

score=r2_score(y_test,regression.predict(x_test) )
print(score)
```
